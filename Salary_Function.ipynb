```python
# ==========================================
# MODULE 2 ASSIGNMENT - SALARY FUNCTION
# Author: Sodiq Omoniyi
# ==========================================

# Import Required Libraries
import pandas as pd
import os
import zipfile

# ------------------------------------------
# STEP 1: IMPORT SALARY DATA
# ------------------------------------------

file_path = os.path.expanduser("~/Downloads/Salary_data.csv")

try:
    salary_data = pd.read_csv(file_path, low_memory=False)

    print("Salary data imported successfully.\n")
    
    print(
        f"✓ Successfully imported {len(salary_data)} records from '{file_path}'"
    )

    print(
        f"✓ Columns: {list(salary_data.columns)}"
    )

except FileNotFoundError:
    print("Salary_data.csv was not found in Downloads folder.")
    raise SystemExit()

# Display first 5 records
print("Dataset Preview:")
print(salary_data.head())

# ------------------------------------------
# STEP 2: PROCESS DATA USING DICTIONARY
# ------------------------------------------

employee_dict = {}

for _, row in salary_data.iterrows():

    employee_dict[str(row["EmployeeName"]).upper()] = {
        "EmployeeName": row["EmployeeName"],
        "JobTitle": row["JobTitle"],
        "BasePay": row["BasePay"],
        "OvertimePay": row["OvertimePay"],
        "OtherPay": row["OtherPay"],
        "Benefits": row["Benefits"],
        "TotalPay": row["TotalPay"],
        "TotalPayBenefits": row["TotalPayBenefits"],
        "Year": row["Year"]
    }

print("\nEmployee dictionary created successfully.")

# ----------------------------------------------------
# STEP 3: CREATE EMPLOYEE FUNCTION AND ERROR HANDLING
# ----------------------------------------------------

def get_employee_details(employee_name):

    try:

        if not isinstance(employee_name, str):
            raise TypeError("Employee name must be a string.")

        employee = employee_dict[employee_name.upper()]

        return employee

    except KeyError:
        return f"Employee '{employee_name}' not found."

    except TypeError as e:
        return str(e)

    except Exception as e:
        return f"Unexpected error: {e}"

# ------------------------------------------
# STEP 4: LOOKUP EMPLOYEE
# ------------------------------------------

employee_name = input("\nEnter Employee Name: ")

employee_details = get_employee_details(employee_name)

print("\nEmployee Details:")
print(employee_details)

# ------------------------------------------
# STEP 5: EXPORT EMPLOYEE DETAILS TO CSV
# ------------------------------------------

if isinstance(employee_details, dict):

    folder_name = "Employee Profile"

    os.makedirs(folder_name, exist_ok=True)

    employee_df = pd.DataFrame([employee_details])

    csv_file = os.path.join(
        folder_name,
        "employee_profile.csv"
    )

    employee_df.to_csv(csv_file, index=False)

    print("\nEmployee profile exported successfully.")

# ------------------------------------------
# STEP 6: ZIP EMPLOYEE PROFILE FOLDER
# ------------------------------------------

    zip_file = "Employee_Profile.zip"

    with zipfile.ZipFile(
        zip_file,
        "w",
        zipfile.ZIP_DEFLATED
    ) as zipf:

        for root, dirs, files in os.walk(folder_name):

            for file in files:

                file_path = os.path.join(root, file)

                zipf.write(
                    file_path,
                    arcname=os.path.relpath(
                        file_path,
                        folder_name
                    )
                )

    print("Employee Profile folder zipped successfully.")
    print(f"ZIP file created: {zip_file}")

else:

    print("Employee not found.")
    print("CSV and ZIP file were not created.")
```
